```python
def mean_distribution(data):
    data.loc[(data['mean'] >= 0) & (data['mean'] <= 3.5), 'mean_group'] = '1'
    data.loc[(data['mean'] > 3.5) & (data['mean'] <= 4.5), 'mean_group'] = '2'    
    data.loc[(data['mean'] > 4.5), 'mean_group'] = '3'
    return data
    
```

```python
code = dataset["loc_code"].values.tolist()
```

```python
code = list(set(code))
```

```python
len(code)
```

```python
df = dataset[["total","loc_code"]]
lo_df_li = []

for i in range(0,len(code)):
    lo_df_li.append(df.loc[df.loc_code == code[i], : ])
```

```python
lo_df_li[2]
```

```python
mean_li = list()
len_li = list()
for each_df in lo_df_li:
    mean_value = each_df["total"].mean()
    mean_li.append(round(mean_value,1))
    len_li.append(len(each_df))
    
```

```python
df1 = pd.DataFrame({
    'loc_code':code,
    'mean':mean_li,
    'number':len_li
})
```

```python
def number_distribution(data):
    data.loc[(data['number'] >= 0) & (data['number'] < 123.3), 'number_group'] = '3'
    data.loc[(data['number'] >= 123.3) & (data['number'] < 246.8), 'number_group'] = '2'    
    data.loc[(data['number'] >= 246.6), 'number_group'] = '1'
    return data
```

```python
def all_distribution(data):
    data.loc[(data['mean_group'] == '1') & (data['number_group'] == '1'), 'all'] = '1'
    data.loc[(data['mean_group'] == '1') & (data['number_group'] == '2'), 'all'] = '2'
    data.loc[(data['mean_group'] == '1') & (data['number_group'] == '3'), 'all'] = '3'
    data.loc[(data['mean_group'] == '2') & (data['number_group'] == '1'), 'all'] = '4'
    data.loc[(data['mean_group'] == '2') & (data['number_group'] == '2'), 'all'] = '5'
    data.loc[(data['mean_group'] == '2') & (data['number_group'] == '3'), 'all'] = '6'
    data.loc[(data['mean_group'] == '3') & (data['number_group'] == '1'), 'all'] = '7'
    data.loc[(data['mean_group'] == '3') & (data['number_group'] == '2'), 'all'] = '8'
    data.loc[(data['mean_group'] == '3') & (data['number_group'] == '3'), 'all'] = '9'
    return data
all_distribution(df1).head()    
```
