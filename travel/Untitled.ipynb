```python
def bar_chart(data,feature,pred) : 
    #ca1 = data[data[pred]==1][feature].value_counts()
    #ca2 = data[data[pred]==2][feature].value_counts()
    ca3 = data[data[pred]==3][feature].value_counts()
    ca4 = data[data[pred]==4][feature].value_counts()
    ca5 = data[data[pred]==5][feature].value_counts()
    df = pd.DataFrame([ca3,ca4,ca5])
    df.index = ['total : 3','total : 4','total : 5']
    df.plot(kind = 'bar', title = feature,stacked = False, figsize=(10,5))
```

```python
ca3 = sati_df[sati_df['total']==3]['munhwa'].value_counts()
ca4 = sati_df[sati_df['total']==4]['munhwa'].value_counts()
ca5 = sati_df[sati_df['total']==5]['munhwa'].value_counts()
df = pd.DataFrame([ca3,ca4,ca5])
df.index = ['total : 3','total : 4','total : 5']
df.plot(kind = 'bar', title = 'munhwa',stacked = True, figsize=(10,5))
```

```python
for i in features:
    bar_chart(sati_df,i,'total')
```
