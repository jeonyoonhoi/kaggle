```python
#12개의 세부 만족도는 X, 9등분한 값은 y
promoting_merged_data = promoting_merged[['nature', 'culture', 'traffic', 'accom', 'food', 'shopping', 'info', 'facility', 'kind', 'program', 'price', 'cong']]
promoting_merged_target = promoting_merged['all']
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
neigh = KNeighborsClassifier(n_neighbors=3, metric='minkowski', p=2)
```

```python
X_train, X_test, y_train, y_test = train_test_split(promoting_merged_data, promoting_merged_target, test_size=0.3)
```

```python
from sklearn.metrics import accuracy_score
```

```python
neigh.fit(X_train, y_train)

y_pred = neigh.predict(X_test)

accuracy_score(y_test, y_pred) 
```

```python
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
from sklearn.model_selection import cross_val_score
```

```python
scores_neigh = cross_val_score(neigh, promoting_merged_data, promoting_merged_target, scoring='accuracy', cv=5)
```

```python
np.mean(scores_neigh), np.std(scores_neigh)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier()

dt.fit(X_train, y_train)

y_pred = dt.predict(X_test)

accuracy_score(y_test, y_pred)
```

```python
scores_dt = cross_val_score(dt, promoting_merged_data, promoting_merged_target, scoring='accuracy', cv=3)
```

```python
np.mean(scores_dt), np.std(scores_dt)
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC(kernel="linear")

svc.fit(X_train, y_train)

y_pred = svc.predict(X_test)

accuracy_score(y_test, y_pred)
```

```python
scores_svc = cross_val_score(svc, promoting_merged_data, promoting_merged_target, scoring='accuracy', cv=5)
```

```python
np.mean(scores_svc), np.std(scores_svc)
```

```python
#recursive feature elimination
#Feature selection 기법 중, wrapper method에 해당
from sklearn.feature_selection import RFE
```

```python
#svc 모델에 영향을 주는 변수 순서대로 랭킹 (돌릴 때 마다 바뀌는데, 3~4개 패턴 있는거 같음)
selector_svc = RFE(svc, n_features_to_select=2)
selector_svc = selector_svc.fit(X_train, y_train)
selector_svc.ranking_
```

```python
#dt 모델에 영향을 주는 변수 순서대로 랭킹 (돌릴 때 마다 바뀌는데, 3~4개 패턴 있는거 같음)
selector_dt = RFE(dt, n_features_to_select=1)
selector_dt = selector_dt.fit(X_train, y_train)
selector_dt.ranking_
```

```python
y_pred = selector_dt.predict(X_test)
```
